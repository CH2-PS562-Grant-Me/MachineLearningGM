```python
import numpy as np
import numpy.ma as ma
import pandas as pd
import tensorflow as tf
from tensorflow import keras
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.model_selection import train_test_split
import tabulate
import random
```

```python
scholarship_url = 'https://raw.githubusercontent.com/qannisa/dummydataset/main/scholarship_features2.csv'

scholarship_file = pd.read_csv(scholarship_url)
```

```python
scholarship_file
```

```python
num_item_features = scholarship_file.shape [1] - 1
item_train = scholarship_file.values.tolist()

item_train = np.array(item_train)
item_train_f = item_train[:, 2:]
item_train_f= np.asarray(item_train_f).astype('float32')
print(item_train_f)
print(item_train_f.shape)

y_train = [row[1] for row in item_train]
y_train = np.asarray(y_train).astype('float32')
y_train.reshape(-1, 1)

print(y_train)
print(y_train.shape)
```

```python
model_item = tf.keras.models.Sequential([
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(64, activation='relu')
])
```

```python
cost_fn = tf.keras.losses.MeanSquaredError()
opt = tf.keras.optimizers.Adam()
model_item.compile(optimizer=opt,
              loss=cost_fn)
```

```python
model_item.fit(item_train_f, y_train, epochs=30)
```

```python
input_item = tf.keras.layers.Input(shape=(3,))
vm = model_item(input_item)
vm = tf.linalg.l2_normalize(vm, axis=1)

print(vm)
```
